```python
#    - 3  2 
import numpy as np 
fruits = np.load('fruits_300.npy')
fruits_2d = fruits.reshape(-1, 100*100)
```

-     np.load()   npy      K-   ( , , )  3  ( , X)   2  . 

-  K-  **sklean.cluster**   **KMeans**   .        **n_clusters**. (   3 .)
-       .    fit()        .  

```python
# KMeans     
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3, random_state=42)
km.fit(fruits_2d)
```

-   KMeans   'labels_'  . 'labels_'     .        . n_clusters=3   'labels_'   0, 1, 2  .

```python
#    
print(km.labels_)
```

-  0, 1, 2     .   0, 1, 2           .    0, 1, 2    .

```python
#  0, 1, 2     
print(np.unique(km.labels_, return_counts=True))
```

-     ( 0) 91  ,   ( 1) 98 ,   ( 2) 111   .             draw_fruits() .

```python
#      draw_fruits()   

import matplotlib.pyplot as plt

def draw_fruits(arr, ratio=1):
    n = len(arr)       # n  
  #  10  .   10     
    rows = int(np.ceil(n/10))
  #  1    .   10.
    cols = n if rows < 2 else 10
    fig, axs = plt.subplots(rows, cols, 
                          figsize=(cols*ratio, rows*ratio), squeeze=False)
    for i in range(rows):
        for j in range(cols):
            if i*10 + j < n:    # n  .
                axs[i, j].imshow(arr[i*10 + j], cmap='gray_r')
            axs[i, j].axis('off')
    plt.show()
```

-  draw_fruits()  ( , , ) 3    10  .        figsize . figsize ratio   . (ratio  1.)
-  2 for         .        . 
-     0    . 

```python
# draw_fruits    0  
draw_fruits(fruits[km.labels_==0])
```

```python
# draw_fruits    1  
draw_fruits(fruits[km.labels_==1])
```

## 4.  K  
- K-          .        .    K    ?
-     K      .       .        **(elbow)**   . 
-    K-           .     **(inertia)** .  '           .'           .             .
-           .            .     .        . 



-     .  KMeans     inertia_  .     K 2 6  KMeans  5 . fit()     inertia_     inertia  .  inertia     . 

```python
#       inertia      
inertia = []
for k in range(2, 7):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(fruits_2d)
    inertia.append(km.inertia_)
plt.plot(range(2, 7), inertia)
plt.xlabel('k')
plt.ylabel('inertia')
plt.show()
```
