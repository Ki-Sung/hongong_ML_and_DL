## 인공지능과 머신러닝, 딥러닝

## 1. 인공지능이란?
### AI(Artificial Intelligence)라고도 불리며, 사람처럼 학습하고 추론할 수 있는 지능을 가진 컴퓨터 시스템을 만드는 기술 

### 1-1. 인공지능의 역사
- 인공지능의 역사는 약 80년 되었지만 오래전부터 지능적인 시스템을 생각했었음.
- 인공지능 역사에서 굵직한 사건으로는 우선 **인공지능 태동기**
    - 1943년 **워런 매컬러**와 **월터 피츠**가 최초로 뇌의 뉴런 개념 발표 (MCP 뉴런)
    - 1950년 **앨런 튜링**이 인공지능이 사람과 같은 지능을 가졌는지 테스트 할 수 있는 **튜링 테스트**를 발표 
    - 1965년 **다트머스 AI 컨퍼런스**개최로 인공지능의 장밋빛 전망이 최고조에 도달함.
- **인공지능 황금기**
    - 1957년 **프랑크 로젠블라트**가 로지스틱 회귀의 기초 버전으로 볼 수 있는 **퍼셉트론**을 발표 
    - 1959년 **데이비드 허블**과 **토르스텐 비셀**이 고양이를 사용하여 시각 피질에 있는 뉴런 기능을 연구
- 하지만 컴퓨터 성능의 한계와 전문가 시스템 구축 실패로 AI의 긴 겨울이 지나가고, 이후 시기를 극복하고 인공지능은 다시 각광받는 분야로 발돋음 함.

### 1-2. 강인공지능(인공일반지능)과 약인공지능
- 강인공지능(인공일반지능): 영화나 드라마, 소설에서 지능을 가진 컴퓨터 시스템이 등장하는 것이 흔함. 흔히 이런 인공 지능을 **강인공지능** 혹은 **인공일반지능** 이라고 부른다. 영화 **그녀** 나 **터미네이터의 스카이넷**이 이 인공지능의 예시로 적합하다.
- 약인공지능: 현재 우리가 마주하고 있는 인공지능을 **약인공지능**이라고 부른다. 대표적인 것이 자율주행, AI 비서로서 사람의 일을 도와주는 보조 역할의 기능만을 함.

## 2. 머신러닝이란?
### 규칙을 일일이 프로그래밍하지 않아도 자동으로 데이터에서 규칙을 학습하는 알고리즘을 연구하는 분야, 인공지능의 하위 분야로 주로 소프트웨어 서비스에 많이 활용된다.

### 2-1. 주요 특징
- 머신러닝은 통계학과 깊은 관련이 있다. 통계학에서 유래된 머신러닝 알고리즘이 많으며 통계학과 컴퓨터 분야가 상호작용하면서 발전하고 있다.
- 컴퓨터 과학 분야에서 대표적인 머신러닝 라이브러리는 **사이킷런**이다. 사이킷런 라이브러리는 **파이썬** API를 사용한다.

## 3. 딥러닝이란?
### 머신러닝 알고리즘 중에 인공 신경망을 기반으로 한 방법들을 통칭하여 딥러닝이라고 부른다. (딥러닝과 인공 신경망을 혼용해서 부른다.)

### 3-1. 주요 특징
- 1998년 **얀 르쿤**이 신경망 모델을 맏늘어 손글씨 숫자를 인식하는데 성공, 이를 합성곱 신경망인 **LeNet-5**라고 부른다.(지금도 쓰고 있음)
- 2012년 **제프리 힌턴**의 팀이 이미지 분류대회인 Image Net에서 기존의 머신러닝 방법을 누르고 압도적인 성능을 선보임. 이 모델을 **AlexNet**이라고 부른다. (이때부터 이미지 분류 작업에 합성곱 신경망이 널리 사용되 시작함.)
- 요즘은 구글의 딥러닝 오픈소스인 **텐서플로**와 페이스 북의 **파이토치**를 많이 사용한다.
- 해당 라이브러리들은 인공신경망 알고리즘을 전문으로 다루고 있고 파이썬 API를 제공함.
